```python
import matplotlib.pyplot as plt

# Plot daily max and min temperature
plt.figure(figsize=(15,6))
df['TMAX'].plot(label='Max Temp', alpha=0.7, color='red')
df['TMIN'].plot(label='Min Temp', alpha=0.7)
plt.title('Daily Temperatures in Richardson (2020–2025)')
plt.ylabel('Temperature (°F)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# As we can see on the graph, there is an outlier in early 2021. This was due to the events of Winte
```

```python
# Top 10 hottest days
df.sort_values('TMAX', ascending=False).head(10)[['TMAX']]
```

```python
# Top 10 coldest days
df.sort_values('TMIN').head(10)[['TMIN']]
```

```python
df['year'] = df.index.year
annual_precip = df.groupby('year')['PRCP'].sum()

plt.figure(figsize=(8,5))
annual_precip.plot(kind='bar', color='skyblue')
plt.title('Annual Precipitation in Richardson')
plt.ylabel('Total Inches')
plt.grid(True)
plt.show()
```

```python
df['month'] = df.index.month
monthly_avg = df.groupby('month')[['TMAX', 'TMIN']].mean()

monthly_avg.plot(kind='bar', figsize=(10,6))
plt.title('Average Monthly Temperatures (2020–2025)')
plt.ylabel('Temperature (°F)')
plt.grid(True)
plt.show()
```
